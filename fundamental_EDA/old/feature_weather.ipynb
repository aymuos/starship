```python
import pandas as pd
```

```python
df = pd.read_csv("package_feature.csv")
df_weather_source = pd.read_csv("weather_data_three_cities.csv")
```

```python
df["receipt_datetime"] = pd.to_datetime(df["receipt_time"])
df["hour"] = df["receipt_datetime"].dt.floor("H")

df = df.merge(
    df_weather_source,
    on=["from_city_name", "hour"],
    how="left"
)
```
