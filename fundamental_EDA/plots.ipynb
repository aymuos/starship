# LIst of plots on the base dataset 



Graph Type	Purpose

----------------------------
- Histogram	: To visualize the distribution of delivery times or the number of packages per delivery.
- Heatmap:	To illustrate the density of delivery requests within different geographical regions.
- Time Series Plot:	To depict trends in delivery demand over time, helping to identify peak periods.
- Scatter Plot:	To explore the relationship between variables such as delivery distance and time taken.
- Bar Chart:	To compare the average delivery times across various cities or zones.
- Box Plot:	To summarize the distribution of delivery times or distances, showing median and quartiles.
- Network Graph:	To visualize interactions and dependencies between different delivery hubs or regions.
- Cumulative Distribution Function (CDF): 	To display the probability of delivery times being less than a certain value.
